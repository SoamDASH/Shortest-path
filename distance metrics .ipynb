```python
import requests
import folium

# Set up the URL and parameters for the API request
url = 'https://api.openrouteservice.org/v2/directions/driving-car'
params = {
    'api_key': '5b3ce3597851110001cf62482425f6e8c4914f3d9fe686b43aa75fdb',
    'start': '8.34234,48.23424',
    'end': '8.34423,48.26424',
    'format': 'geojson'
}

# Send the API request and retrieve the response
response = requests.get(url, params=params)

# Extract the distance and duration from the response
data = response.json()
distance = data['features'][0]['properties']['segments'][0]['distance']
duration = data['features'][0]['properties']['segments'][0]['duration']

# Print the distance and duration
print(f"Distance: {distance} meters")
print(f"Duration: {duration} seconds")

route_coords = data['features'][0]['geometry']['coordinates']


# Create a folium map centered on the start location
start_location = route_coords[0][::-1]
m = folium.Map(location=start_location, zoom_start=10)

# Create a folium polyline from the route coordinates and add it to the map
polyline = folium.PolyLine(locations=[coord[::-1] for coord in route_coords])
polyline.add_to(m)

folium.TileLayer('https://tiles.waqi.info/tiles/usepa-aqi/{z}/{x}/{y}.png?token=_TOKEN_', name='traffic', attr='Air Quality Index').add_to(m)


# Display the map
m
```
